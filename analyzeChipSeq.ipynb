```python
import pandas as pd
from TF import *
```

```python
result  = pd.DataFrame()
for ind, TF in enumerate(TFs):
    print('Reading data for TF ', TF, ' (', ind+1,'/10)')
    data = pd.read_excel('chipSeq2.xlsx', header=0, sheet_name=TFsimple[ind])
    data = data[data['Chromosome'] == 'chr8']
    data['TF'] = TF
    try:
        data = data.drop(columns=['pValue', 'qValue', 'score', 'signalValue', 'peak.1', 'name', 'strand', 'length of segment'])
    except:
        data = data.drop(columns=['pValue', 'qValue', 'score', 'signalValue', 'name', 'strand', 'length of segment'])
    if ind == 0:
        result = data
    else:
        result = result.append(data)

result['peakPosition'] = result['chromStart'] + result['peak']
result.head(-1)
```

```python
# save final dataframe to pickle file
result.to_pickle('ChIPseq_chr8_final10_low.pkl')
```
